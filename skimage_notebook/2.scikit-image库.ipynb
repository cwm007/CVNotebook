# 6.图像处理

## 6.1 腐蚀和膨胀操作

腐蚀(erosion)和膨胀(dilation)操作是对于白色部分（高亮部分），而不是黑色部分。腐蚀操作会让高亮部分区域更小，而膨胀操作会使高亮部分区域更大。膨胀是求局部最大值的操作，而腐蚀是求局部最小值的操作。

```python
from skimage import data,io,morphology

# 加载灰度图
img = data.checkerboard()
io.imshow(img)
io.show()
```

```python
# 腐蚀操作
erosion_img = morphology.erosion(img, morphology.square(8))
io.imshow(erosion_img)
io.show()
```

```python
# 膨胀操作
dilation_img = morphology.dilation(img, morphology.square(10))
io.imshow(dilation_img)
io.show()
```

## 6.2 自定义卷积核

```python
from PIL import ImageFilter,Image
import numpy as np

my_kernel = ImageFilter.Kernel((3,3),np.arange(9))
img = Image.open('../example_images/cat.jpg')
img = img.filter(my_kernel)
img.show()
```
