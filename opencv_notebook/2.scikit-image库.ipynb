# 2.图像格式

常见的图像格式有三种：JPG/JPEG(Joint Photographic Experts Group)、PNG(Portable Network Graphics)、GIF(Graphics Interchange Format)，JPG和JPEG格式是相同的，没有区别。PNG和GIF格式图像特征损失较少，而JPG图像特征损失较大。

# 3.图像基本操作

## 3.1 读取和展示图像

```python
from skimage import io
img = io.imread('../example_images/cat.jpg')
io.imshow(img)
io.show()
```

注意：skimage库展示图像时会将像素点的坐标给展示出来。

## 3.2 保存图像

```python
io.imsave('../example_images/copy_cat2.jpg',img)
```
